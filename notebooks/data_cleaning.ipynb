```python
# Convert datatype 
categorical_cols= ["brand_name","category_name","sub_category_name","product_type"
                 ,"status","is_approved"] 
datetime_cols = ["offer_start_price","offer_end_price","created_at"]

text_cols = ["product_name",
    "slug",
    "short_description",
    "long_description",
    "sku",
    "seo_title",
    "seo_description"]
for col in datetime_cols: 
    df_clean[col] = pd.to_datetime(df_clean[col]) 
for col in categorical_cols: 
    df_clean[col] = df_clean[col].astype("category") 
for col in text_cols: 
    df_clean[col] = df_clean[col].astype(str).str.strip().str.lower() 
    
```

```python
# Check text data: 
import re
def remove_html_regex(text): 
    if pd.isna(text): return "Unavailable" 
    clean = re.sub(r"<.*?>", "", text) 
    return clean 
df_clean["short_description"] = df_clean["short_description"].apply(remove_html_regex)
df_clean["long_description"] = df_clean["long_description"].apply(remove_html_regex)
for col in text_cols: 
    df_clean[col] = df_clean[col].str.strip().str.lower().str.replace(r"\s+", " ", regex=True)
```

```python
#Check category
for col in categorical_cols: 
    print(df_clean[col].unique())
```

```python
#Export Cleaned data 
df_clean.to_csv("cleaned_products.csv",index=False) 
```

```python
df_clean.to_parquet("cleaned_products.parquet",index=False)
```
